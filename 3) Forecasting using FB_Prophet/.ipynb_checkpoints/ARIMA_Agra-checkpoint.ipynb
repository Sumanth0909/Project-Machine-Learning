```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load your dataset
data = pd.read_csv('state_weather.csv')

# Convert 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Sort the data by date
data = data.sort_values('date')

# Select the column for prediction (e.g. 'PM2.5')
ts_data = data[['date', 'PM2.5']]
ts_data.set_index('date', inplace=True)

# Train-test split
train_data = ts_data[:int(0.8*(len(ts_data)))]
test_data = ts_data[int(0.2*(len(ts_data))):]

# Fit ARIMA model
model = ARIMA(train_data, order=(5,1,0))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=len(test_data))

# Visualize the results
plt.figure(figsize=(12,6))
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Testing Data')
plt.plot(test_data.index, forecast, label='Forecast', color='black')
plt.legend()
plt.title('ARIMA Forecasting')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Drop rows with missing values
test_data = test_data.dropna()
forecast = forecast.dropna()

# Align the lengths of test_data and forecast
min_length = min(len(test_data), len(forecast))
test_data = test_data[:min_length]
forecast = forecast[:min_length]

# Calculate Mean Squared Error
mse = mean_squared_error(test_data, forecast)
print("Mean Squared Error:", mse)

# Calculate R-squared score
r2 = r2_score(test_data, forecast)
print("R-squared Score:", r2)
```
