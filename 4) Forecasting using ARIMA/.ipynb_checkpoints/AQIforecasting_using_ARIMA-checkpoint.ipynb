```python
#  Find unique locations
unique_locations = data['location'].unique()
print("Unique locations:")
for i, location in enumerate(unique_locations, 1):
    print(f"{i}. {location}")



def determine_data_type(user_input):
    # Attempt to convert to integer
    try:
        int_value = int(user_input)
        return 'integer'
    except ValueError:
        pass
    
    # Attempt to convert to float
    try:
        float_value = float(user_input)
        return 'float'
    except ValueError:
        pass
    
    # If both conversions fail, it remains a string
    return 'string'

# Prompt the user to select a location
selected_index = input("Select a location by number: ")


# Determine the data type of the input
data_type = determine_data_type(selected_index)


if data_type == 'integer' and  int(selected_index)>=1 and int(selected_index)<=304:
    
    selected_index = int(selected_index) -1
    selected_location = unique_locations[selected_index]
    print(f"Selected location: {selected_location}")

    #  Filter the data based on the selected location
    filtered_data = data[data['location'] == selected_location]

    # Display the first few rows of the filtered data
    print(filtered_data.head())
    
else :
    print("Please enter a valid number to determine the location!")
```

```python
# Drop rows with missing values in 'date'
data_cleaned = filtered_data.dropna(subset=['date'])

# Create a new DataFrame for ARIMA
arima_df = data_cleaned[['date', 'AQI_calculated']].rename(columns={'date': 'ds', 'AQI_calculated': 'y'})

# Check for missing values in the new DataFrame
missing_values_arima_df = arima_df.isnull().sum()
print("Missing values in arima_df:\n", missing_values_arima_df)
```

```python
# Display the first few rows of the new DataFrame
print(arima_df.head())
```

```python
arima_df.isnull().sum()
```

```python
# For choosing the value of 'd'



from statsmodels.tsa.stattools import adfuller

stationarity_test_result=adfuller(arima_df['y'])
print('ADF Statistic:', stationarity_test_result[0])
print('p-value:', stationarity_test_result[1])

# If p-value > 0.05, the series is non-stationary and you may need to difference it


if stationarity_test_result[1] > 0.05 :
    print("The series is non-stationary and you may need to difference it")
    
else :
    print("The series is significantly stationary. So, there is no need to difference it. Also, the order of differencing = 0")
    d = 0
```

```python
# For choosing the value of 'p' and 'q'


# we will use "auto_arima" for Automated Selection 
# The auto_arima function from pmdarima automates the selection of 'p','d','q'
# Since we have already decided that d = 0 for our case, we will choose only 'p' and 'q' using "auto_arima"
```

```python
from pmdarima import auto_arima


ts = arima_df['y']

# Auto ARIMA to find the best p, d, q
model = auto_arima(ts, start_p=1, start_q=1,
                   max_p=5, max_q=5, seasonal=False,
                   d=0, trace=True,                            # since its decided that d = 0
                   error_action='ignore', suppress_warnings=True, 
                   stepwise=True)

print(model.summary())
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
fig, axes = plt.subplots(1, 2, figsize=(16, 4))
plot_acf(ts, lags=40, ax=axes[0])
plot_pacf(ts, lags=40, ax=axes[1])
plt.show()
```

```python
# define 'p','d' and 'q'

p = 2
d = 0   # already set
q = 0
```

```python
# Split into train and test sets
train_size = int(len(arima_df) * 0.8)  # 80% train, 20% test
train_df = arima_df.iloc[:train_size]
test_df = arima_df.iloc[train_size:]

print("\nTraining DataFrame:")
print(train_df)
print("\nTesting DataFrame:")
print(test_df)
```

```python
                                        # Training the ARIMA model

from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model (adjust p, d, q as determined from ACF/PACF)
arima_model = ARIMA(train_df['y'], order=(p, d, q))
model_fit = arima_model.fit()

# Print confirmation
print("Model training complete.")

print(model_fit.summary())
```

```python
                                    # Testing the ARIMA model

# Define the start and end of the forecast period
start = len(arima_df)
end = start + len(test_df) - 1

# Generate forecast
forecast = model_fit.predict(start=start, end=end, dynamic=True)

# Create a DataFrame for the forecast
forecast_test = pd.DataFrame({'ds': test_df.index, 'yhat': forecast})
```

```python
                                    # Evaluating the ARIMA model


from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error


# Merge actual values with predicted values
comparison_df = test_df[['ds','y']].merge(forecast_test[['ds', 'yhat']], left_index=True, right_on='ds')

# Calculate MAE
mae = mean_absolute_error(comparison_df['y'], comparison_df['yhat'])
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate MAPE
mape = mean_absolute_percentage_error(comparison_df['y'], comparison_df['yhat'])
print(f'Mean Absolute Percentage Error (MAPE): {mape}')
```

```python
# Calculate MAE and MAPE for each year in the dataset
years = arima_df['ds'].dt.year.unique()
mae_per_year = []
mape_per_year = []

for year in years:
    yearly_data = arima_df[arima_df['ds'].dt.year == year]
    if len(yearly_data) < 2:  # Skip years with insufficient data for ARIMA modeling
        continue
    
    # Split into train and test sets
    train_size = int(len(yearly_data) * 0.8)  # 80% train, 20% test
    train_df = yearly_data.iloc[:train_size]
    test_df = yearly_data.iloc[train_size:]

    # Fit the ARIMA model
    arima_model = ARIMA(train_df['y'], order=(p, d, q))
    model_fit = arima_model.fit()

    # Generate forecast
    start = len(train_df)
    end = start + len(test_df) - 1
    forecast = model_fit.predict(start=start, end=end, dynamic=True)

    # Merge actual values with predicted values
    forecast_df = pd.DataFrame({'yhat': forecast})
    forecast_df.index = test_df.index[:len(forecast)]  # Ensure the forecast index matches the test data index
    comparison_df = test_df[['ds', 'y']].join(forecast_df)

    # Drop rows with NaN values
    comparison_df = comparison_df.dropna()

    # Calculate MAE and MAPE
    mae = mean_absolute_error(comparison_df['y'], comparison_df['yhat'])
    mape = mean_absolute_percentage_error(comparison_df['y'], comparison_df['yhat'])
    
    mae_per_year.append(mae)
    mape_per_year.append(mape)

# Plot MAE and MAPE against year
plt.figure(figsize=(14, 6))

# Plot MAE
plt.subplot(1, 2, 1)
plt.plot(years, mae_per_year, marker='o')
plt.xlabel('Year')
plt.ylabel('Mean Absolute Error (MAE)')
plt.title('MAE vs Year')

# Plot MAPE
plt.subplot(1, 2, 2)
plt.plot(years, mape_per_year, marker='o')
plt.xlabel('Year')
plt.ylabel('Mean Absolute Percentage Error (MAPE)')
plt.title('MAPE vs Year')

plt.tight_layout()
plt.show()
```

```python
                                    # Interactive Prediction

# Ask user for input
user_time = input("Please enter the time period for which you want the prediction (in number of days): ")

# Determine the data type of the input
data_type = determine_data_type(user_time)

if data_type == 'integer' and int(user_time) >= 0:
    # Generate forecast for the specified number of days
    future_periods = int(user_time)
    start = len(arima_df)
    end = start + future_periods - 1
    
    # Make predictions
    forecast1 = model_fit.predict(start=start, end=end, dynamic=True)
    
    # Adjust future_dates to match the length of forecast1
    future_dates = pd.date_range(start=arima_df['ds'].max() + pd.Timedelta(days=1), periods=future_periods)
    forecast1_df = pd.DataFrame({'ds': future_dates, 'yhat': forecast1[:future_periods]})
    
    # Display the forecast
    print(forecast1_df[['ds', 'yhat']].head())
    
    # Plot the forecast
    plt.figure(figsize=(10, 5))
    plt.plot(arima_df['ds'], arima_df['y'], label='Actual')
    plt.plot(forecast1_df['ds'], forecast1_df['yhat'], label='Forecast')
    plt.xlabel('Date')
    plt.ylabel('AQI')
    plt.title('ARIMA Forecast')
    plt.legend()
    plt.show()
    
else:
    print("Please enter a valid number greater than or equal to 0!")
```
